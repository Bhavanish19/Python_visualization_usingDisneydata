## EDA

```python
type_counts = data['type'].value_counts()

bar_width = 0.8


plt.figure(figsize=(6, 2))

# Plot the counts for movies and TV shows
plt.barh(type_counts.index, type_counts.values, color=['skyblue', 'violet'])  # Customize colors as needed


plt.xlabel('Count')
plt.ylabel('Type')
plt.title('Number of Movies and TV Shows on Disney Plus')


plt.show()
```

```python
comedy_count = data['listed_in'].str.contains('Comedy', case=False, na=False).sum()
action = data['listed_in'].str.contains('Action-Adventure', case=False, na=False).sum()
Family = data['listed_in'].str.contains('Family', case=False, na=False).sum()
Animation = data['listed_in'].str.contains('Animation', case=False, na=False).sum()
Documentary = data['listed_in'].str.contains('Documentary', case=False, na=False).sum()
values = [comedy_count, action, Family, Animation, Documentary]
labels = ["comedy_count", "action", "Family", "Animation", "Documentary"]
plt.figure(figsize=(8, 8))  # Adjust the figure size as needed
plt.pie(values, labels=labels, autopct='%1.1f%%', startangle=120, colors=["skyblue", "indigo", "green", "pink", "violet"])

# Add a title
plt.title('Pie chart showing common listed_in categories')

# Show the chart
plt.axis('equal')  # Ensures the pie chart is a circle
plt.show()
```

```python
sns.set_style("darkgrid", {"grid.color": ".5", "grid.linestyle": ":"})
sns.set_context('notebook')
g=sns.catplot(x='rating',kind='count',data=data,col='type',height=5)
g.set_xticklabels(rotation=90)
g.set(xlabel='The rating',ylabel='The count of each rating')
g.fig.suptitle('Number of movies and shows per type and rating',y=1.03)
g.set_titles('{col_name}s')

plt.show()
```

```python
s_m_per_year=data.groupby('release_year').size()
s_m_df=s_m_per_year.reset_index()
s_m_df.columns=['year','frequency']
g=sns.relplot(data=s_m_df, x='year',y='frequency',kind='line',color='b',height=9)
g.fig.suptitle('Number of Movies and Shows Released Per Year',y=1.04)
g.set(xlabel='Release Year',ylabel='Number of Movies and Shows')
plt.show()
```
